```python
class PDF(FPDF):
    def header(self):
        # logo
        #self.image("logo.png",10,8,25)
        
        #Font
        self.set_font("times","b",20)
        
        #Color of frame
        self.set_draw_color(0,80,180)
        #Set fill color
        self.set_fill_color(230,230,50)
        #Set text color
        self.set_text_color(220,50,50)
        #Title
        self.cell(0,10,"Room Cleaning Report",border=True,ln=1,align="C")
        self.ln(20)
    def footer(self):
        self.set_y(-15)
        self.set_font("helvetica","I",8)
        self.cell(0,10,"Page Number {page_number}".format(page_number=self.page_no()),border=False,ln=1,align="C")
        
pdf = PDF("L","mm","A4")
pdf.add_page()
pdf.set_auto_page_break(auto=True,margin=10)
pdf.set_font("Times", size=10)
line_height = 25
#col_width = pdf.epw / product_frame.shape[1] 


##Seeting links for Pages
prd_link  = pdf.add_link()
prd_link1 = pdf.add_link()


pdf.set_font("times","b",10)
pdf.cell(0,10,"Product Details",link=prd_link,ln=True)
pdf.cell(0,10,"Product Details 11",link=prd_link1,ln=True)
pdf.add_page()



############################################# Start of Product Details Table #######################################

pdf.set_link(prd_link)
pdf.set_font("times","b",10)
pdf.cell(0,10,"Product Details",border=False,ln=True,align="L")
pdf.ln(5)

#pdf.create_table(table_data=product_frame.values.tolist(),title="Product Details",cell_width="even",align_header="R",data_size=8)

pdf.multi_cell(0, line_height, "Product Details", border=0, align='C', ln=3, max_line_height=8)
pdf.ln(line_height)
pdf.set_font("Times",size=10)
pdf.multi_cell(20, line_height, "Generic Name", border=1, ln=3,max_line_height=12)
pdf.multi_cell(20, line_height, "Form", border=1, ln=3,max_line_height=12)
pdf.multi_cell(20, line_height, "API with strength", border=1, ln=3,max_line_height=12)
pdf.multi_cell(20, line_height, "Minimum Batch size NOS", border=1, ln=3,max_line_height=12)

pdf.multi_cell(20, line_height, "Minimum Batch size MG", border=1, ln=3,max_line_height=12)
pdf.multi_cell(15, line_height, "MRDD", border=1, ln=3,max_line_height=12)
pdf.multi_cell(15, line_height, "LRDD MG", border=1, ln=3,max_line_height=12)


pdf.multi_cell(15, line_height, "LRDD NOS", border=1, ln=3,max_line_height=12)
pdf.multi_cell(15, line_height, "PDE VALUE", border=1, ln=3,max_line_height=12)
pdf.multi_cell(15, line_height, "LD50", border=1, ln=3,max_line_height=12)
pdf.multi_cell(15, line_height, "NOEL", border=1, ln=3,max_line_height=12)

pdf.ln(line_height)

pdf.set_font("Times",size=8)
line_height = pdf.font_size * 2.5
for row in product_frame.itertuples():
    pdf.multi_cell(20, line_height, row.Generic_Name, border=0, ln=3)
    pdf.multi_cell(20, line_height, row.Form, border=0, ln=3)
    pdf.multi_cell(20, line_height, row.API_with_strength, border=0, ln=3)
    pdf.multi_cell(20, line_height, row.Minimum_Batch_size_NOS, border=0, ln=3)
    
    pdf.multi_cell(20, line_height, row.Minimum_Batch_size_MG, border=0, ln=3)
    pdf.multi_cell(15, line_height, row.MRDD, border=0, ln=3)
    pdf.multi_cell(15, line_height, row.LRDD_MG, border=0, ln=3)
    
    
    pdf.multi_cell(15, line_height, row.LRDD_NOS, border=0, ln=3)
    pdf.multi_cell(15, line_height, row.PDE_VALUE, border=0, ln=3)
    pdf.multi_cell(15, line_height, row.LD50, border=0, ln=3)
    pdf.multi_cell(15, line_height, row.NOEL, border=0, ln=3)
    pdf.ln(line_height)

    
    
# ############################################# End of Product Details Table #######################################


pdf.output("example.pdf")
```
