```python
import mysql.connector
from mysql.connector import errorcode
config1 = {
  'host':"pinpointserver.mysql.database.azure.com",
  'port':"3306",
  'user':"arissdb",
  'password':"Ashijyoti@12",
  'db':'ROOMCLEANINGDB'
}


try:
    conn1 = mysql.connector.connect(**config1)
    print("Connection established")
except mysql.connector.Error as err:
    if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
        print("Something is wrong with the user name or password")
    elif err.errno == errorcode.ER_BAD_DB_ERROR:
        print("Database does not exist")
    else:
        print(err)
else:
    cursor1 = conn1.cursor()
```

```python
cursor1.execute(""" drop table if exists  ROOMCLEANINGDB.USERS """)
conn1.commit()

cursor1.execute(""" create table ROOMCLEANINGDB.USERS 
 (USERNAME varchar(255), FNAME varchar(255),
 LNAME varchar(255), ROLE varchar(255), PASSWORD varchar(255), EMAILID varchar(255),
 STATUS varchar(255) , photo_blob BLOB)""")
conn1.commit()
```

```python
cursor1.execute(""" drop table if exists  ROOMCLEANINGDB.product_Details """)
conn1.commit()

cursor1.execute(""" create table ROOMCLEANINGDB.product_Details 
 (Product_Name varchar(255),
 Generic_Name varchar(255),
 Form varchar(255), 
 API_with_strength varchar(255), 
 Minimum_Batch_size_NOS varchar(255), 
 Minimum_Batch_size_MG varchar(255),
 MRDD varchar(255) , 
 LRDD_MG varchar(255) , 
 LRDD_NOS varchar(255) , 
 PDE_VALUE varchar(255) , 
 LD50 varchar(255) , 
 NOEL varchar(255),
 VERSION varchar(255) , 
 STATUS varchar(255) , 
 UPDATED_BY varchar(255) ,
 `date` TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP)""")
conn1.commit()

# Product_Name','Generic_Name','Form', 'API_with_strength' ,'Minimum_Batch_size_NOS',
#                        'Minimum_Batch_size_MG','MRDD','LRDD_MG','LRDD_NOS','PDE_VALUE','LRD50','NOEL'

cursor1.execute(""" drop table if exists  ROOMCLEANINGDB.file_repo """)
conn1.commit()

cursor1.execute(""" create table ROOMCLEANINGDB.file_repo  (
             report_name varchar(255),
             prepared_by varchar(255),
             file_blob BLOB,
             `date` TIMESTAMP NOT NULL DEFAULT CURRENT_TIMESTAMP )""")
conn1.commit()
```
